```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data = pd.read_csv('creditcard.csv')
# data.hist()
# plt.show()
print 'done'
```

```python
pd.value_counts(data.Class).plot.bar()
plt.title('Histogram of Class')
plt.ylabel('frequency')
plt.xlabel('class')
plt.show()
print
print data.Class.value_counts()
```

```python

pos = 0
header = data.columns.tolist()
print header
# univariate visualisation
# for column in data:
#     print data[column]
#     break
for column in data:
    
    #plt.hist(x=data[column], bins='auto', color='#0504aa', alpha=0.9, rwidth=0.85, histtype='bar')
    #plt.show()
    plt.hist(x=data[column], bins='auto')
    
    plt.xlabel(header[pos])
    plt.ylabel("Frequency")
    plt.title("Histogram of "+header[pos])
    plt.show()
    pos += 1
    #if pos == 4:
    #    break
    #break
```
